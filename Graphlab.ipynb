```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

# Load the data from elastic

```python
from elasticsearch import Elasticsearch
es = Elasticsearch(host='atteridgeville.csir.co.za',port=9200)
```

```python
l = {
    "query": { "match_all": {} },
    "size": 490
         
    }
```

```python
res = es.search(index="khulo_final", doc_type="web", body=l)
```

```python
len(res)
```

```python
descriptions=[]
titles = []
sal =[]
for i in range (0,len(res['hits']['hits'])):
    y = res['hits']['hits'][i]['_source']['unclean_desc']
    descriptions.append(y)
```

```python
len(descriptions)
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
descriptions[2]
```

```python
num = len(descriptions)
for i  in range(len(descriptions)):
    descriptions[i] =[stemmer.stem(word) for word in descriptions[i].split()]
    descriptions[i] = ' '.join(descriptions[i])
```

```python
descriptions[0]
```

```python
for i in range(len(titles)):
    if "Information Technology" in titles[i]:
        titles[i] = titles[i].split('-',1)[1:]
    else:
        titles[i]= titles[i]
```

## Create the module

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',stop_words = 'english',
                                lowercase = True,token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, min_df = 10)
dtm_tf = tf_vectorizer.fit_transform(descriptions)
print(dtm_tf.shape)
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(descriptions)
print(dtm_tfidf.shape)
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_topics=10,random_state=1)
lda_tf.fit(dtm_tf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_topics=10, random_state=1)
lda_tfidf.fit(dtm_tfidf)
```

```python
lda_tf.perplexity(dtm_tf)
```

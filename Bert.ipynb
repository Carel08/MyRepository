```
#Standardize sublists 
def review_to_words(raw_review):
    review_text=BeautifulSoup(raw_review).get_text() #Removes unwanted characters
    letters_only=re.sub("[^a-zA-Z0-9]", " ", review_text) #Removes numbers and symbols
    words=letters_only.lower().split() #puts everything in lowercase
    stops=set(stopwords.words("english")) #removes stopwords (and, or...)
    meaningful_words=[w for w in words if not w in stops] #Keeps only meaningful words in a list
    return (" ".join(meaningful_words))
```
